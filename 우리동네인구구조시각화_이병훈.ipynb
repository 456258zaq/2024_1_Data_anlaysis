```python
import csv

file = open("202403_202403_연령별인구현황_월간 (2) (1).csv",encoding="cp949")

dataset = csv.reader(file)
result = []
for i in dataset:
  if '은행동' in i[0]:
    print(i)
    for j in i[3:]:
      result.append(int(j.replace(",","")))

file.close()
```

```python
len(result)
```

```python
import matplotlib.pyplot as plt
plt.plot(result)
plt.rc('font', family="NanumBarunGothic")
plt.title("은행동 인구구조")
plt.plot(result,color="magenta",label="인구수")
# plt.grid(visible=True)
plt.xlabel("나이")
plt.ylabel("인구수")
plt.legend()
plt.show()
```
